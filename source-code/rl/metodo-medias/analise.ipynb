```python
import pandas as pd
import matplotlib.pyplot as plt

# Configuração de estilo
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({'font.size': 12})

# 1. Carregar os dados
csv_file = 'average_methods_results.csv'
try:
    df = pd.read_csv(csv_file)
except FileNotFoundError:
    print(f"Erro: Arquivo '{csv_file}' não encontrado. Execute a simulação C++ primeiro.")
    exit(1)

# 2. Criar o gráfico
plt.figure(figsize=(10, 6))

# Plotar as curvas
# Nota: Com a sua modificação no C++, ambas as colunas podem estar idênticas
# se a função run_simulation for determinística e chamada com os mesmos parâmetros.
plt.plot(df['step'], df['greedy_reward'], 
         label='Método Modificado (Alpha=0.9)', 
         color='blue', linewidth=2)

plt.plot(df['step'], df['epsilon_greedy_reward'], 
         label='Comparativo (Se houver diferença)', 
         color='red', linestyle='--', linewidth=2, alpha=0.7)

# 3. Configurações dos Eixos e Legenda
plt.title('Performance de Recompensa Acumulada', fontsize=14, pad=15)
plt.xlabel('Interações (Tempo)', fontsize=12)
plt.ylabel('Recompensa Total Acumulada', fontsize=12)
plt.legend(loc='upper left', frameon=True)
plt.grid(True, linestyle='--', alpha=0.7)

# 4. Salvar e Mostrar
output_file = 'comparativo_performance.pdf'
plt.tight_layout()
plt.savefig(output_file)
print(f"Gráfico salvo como '{output_file}'")

# plt.show() # Descomente se quiser visualizar na tela
```
